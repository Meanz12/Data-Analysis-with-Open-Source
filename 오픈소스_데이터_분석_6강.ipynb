# 오픈소스 기반 데이터 분석 6강 - 데이터 전처리 1

```python
import requests

## params
api_key = "6a4e5366496d656135315452646c4a"
api_type = 'json'
api_service = 'energyUseDataSummaryInfo'
api_startIndex = 1
api_endIndex = 5

## needed info
needed_columns = ['YEAR', 'MON', 'MM_TYPE', 'EUS', 'GUS', 'WUS', 'HUS']

## function
def get_year_months():
    year_months = []
    for year in range(2015, 2024+1):
        for month in range(1,12+1):
            year_months.append(f'{year}{month:02d}')
    return year_months

year_months = get_year_months()
results = []

for ym in get_year_months():
    year = ym[:4]
    month = ym[4:]
    url = f"http://openapi.seoul.go.kr:8088/{api_key}/{api_type}/{api_service}/{api_startIndex}/{api_endIndex}/{year}/{month}"
    response = requests.get(url, timeout=3)

    if response.status_code == 200:
        #print("api 호출 성공")
        data = response.json()
        rows = data[api_service]['row']
        filtered_rows = []

        for row in rows:
            if row['MM_TYPE'] == "개인":
                new_row = {}
                for col in needed_columns:
                    new_row[col] = row[col]
                filtered_rows.append(new_row)
        results.extend(filtered_rows)
    else :
        print(f"{year}{month}: {response.status_code} 오류")

print(results)
```

```python
import pandas as pd

df = pd.DataFrame(results)
## DataFrame 인덱스 출력
print(df.index)
## DataFrame 컬럼 출력
print(df.columns)
## DataFrame 행 출력
print(df.values.tolist())
print(df.values) # python numpy
## DataFrame 값 출력
print(df.values.flatten())
```

```python

# 계절 구분 함수
def get_season(month):
    if month in [3, 4, 5]:
        return '봄'
    elif month in [6, 7, 8]:
        return '여름'
    elif month in [9, 10, 11]:
        return '가을'
    else:
        return '겨울'

# 계절 컬럼 추가
df['season'] = df['month'].apply(get_season)

print("전처리 완료 — year, month, season 컬럼 추가됨:")
df
```
